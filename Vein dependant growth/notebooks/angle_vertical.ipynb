```python
def theta(growth_front):
    
    theta_front = []
    u, v = list(growth_front[3,:]), list(growth_front[4,:])
    
    for i in range(growth_front.shape[1]):
        if v[i] >= 0:
            theta_front.append(math.atan(u[i] / v[i]))
            
        else:
            if u[i] >= 0:
                theta_front.append(math.atan(u[i] / v[i]) + math.pi )
            else:
                theta_front.append(math.atan(u[i] / v[i]) - math.pi )
    
    return theta_front
```
